# Guia practica: analisis de sensibilidad del sistema de Lorenz

Este notebook reune el analisis ampliado de sensibilidad para el sistema de Lorenz con `RK4`, incluyendo sensibilidad a condiciones iniciales, barrido fino en `r`, sensibilidad a `dt`, transitorio y observables resumidos.

```python
import json
from pathlib import Path
import matplotlib.pyplot as plt

BASE_DIR = Path.cwd().resolve().parents[0]
SUMMARY_PATH = BASE_DIR / 'content' / 'files_lorenz_sensibilidad' / 'resumen_sensibilidad_lorenz.json'
with SUMMARY_PATH.open('r', encoding='utf-8') as f:
    summary = json.load(f)

summary.keys()
```
